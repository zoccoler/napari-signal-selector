```python
from matplotlib.lines import Line2D
from matplotlib.pyplot import scatter
import matplotlib.pyplot as plt
import numpy as np
from napari_signal_selector.utilities import get_custom_cat10based_cmap_list
from matplotlib.colors import ListedColormap, Normalize
```

```python
cmap = get_custom_cat10based_cmap_list()
```

```python
len(cmap)
```

```python
my_mpl_cmap = ListedColormap(cmap)
my_mpl_cmap
```

```python
normalizer = Normalize(vmin=0, vmax=len(cmap) - 1)
normalizer
```

```python
# Normalization is messing with numbers in 'c' that should be indices for colors
x = np.arange(10)
y = np.random.rand(10)
plt.scatter(x, y, c=[5,1,1,1,1,2,2,2,2,4], cmap=my_mpl_cmap, norm=normalizer)
```
